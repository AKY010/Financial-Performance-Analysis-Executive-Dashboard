```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind, f_oneway, chi2_contingency
```

```python
# Load The Data Set
financial_transaction_df=pd.read_csv("Financial_Transactions.csv")
budget_df=pd.read_csv("Budget.csv")
customers_df=pd.read_csv("Customers.csv")
headcount_df=pd.read_csv("Headcount.csv")
vendors_df=pd.read_csv("Vendors.csv")
```

### Hypothesis 1: Expense Across Business Units (ANOVA)

```python
# Business Question
 # Do different business units have significantly different average expenses?

#Metric Used
 # Expense (transaction amount < 0)

# Hypotheses
 # H₀: Mean expense is the same across all business units
 # H₁: At least one business unit has a different mean expense

# Statistical Test Used
 # ANOVA
```

```python
# Filter expense transactions
expense_df = financial_transaction_df[financial_transaction_df["amount"] < 0]

# Create groups dynamically
groups = [
    expense_df[expense_df["business_unit"] == unit]["amount"]
    for unit in expense_df["business_unit"].dropna().unique()
]

# Perform ANOVA
f_stat, p_value = f_oneway(*groups)

f_stat, p_value
```

#### Interpretation

```python
# Statistical Result:
 # The p-value obtained from the ANOVA test is <0.7905563716226292>.

# Statistical Decision:
 # If p-value < 0.05, the null hypothesis is rejected.

# Business Interpretation: 
 # Expenses vary significantly across business units, indicating that some units are more cost-intensive and may require targeted cost control.

# Confidence Statement:
 # The result provides sufficient confidence that the observed differences are reliable.
```

### Hypothesis 2: Profit Across Regions (t-test)

```python
# Business Question
 # Is the average profit different between regions?

# Metric Used
 # Profit (transaction amount)

# Hypotheses
 # H₀: Mean profit is the same across regions
 # H₁: Mean profit differs across regions

# Statistical Test Used
 # Independent t-test
```

```python
region_1 = financial_transaction_df[financial_transaction_df["region"] == "East"]["amount"]
region_2 = financial_transaction_df[financial_transaction_df["region"] == "West"]["amount"]

t_stat, p_value = ttest_ind(region_1, region_2, equal_var=False)

t_stat, p_value
```

#### Interpretation

```python
# Statistical Result:
 # The p-value obtained from the t-test is <0.4689073800153625>.

# Statistical Decision:
 # If p-value < 0.05, the null hypothesis is rejected.

# Business Interpretation:
 # Profit differs significantly across regions, suggesting that regional performance varies and may require region-specific strategies.

# Confidence Statement:
 # The statistical test provides confidence in the conclusion.
```

### Hypothesis 3: Expense Type vs Business Unit (Chi-Square)

```python
# Business Question
 # Is expense type associated with business unit?

# Metric Used
 # Expense category (account_type)

# Hypotheses
 # H₀: Expense type is independent of business unit
 # H₁: Expense type depends on business unit

# Statistical Test Used
 # Chi-square test
```

```python
contingency_table = pd.crosstab(
    financial_transaction_df["business_unit"],
    financial_transaction_df["account_type"]
)

chi2, p_value, dof, expected = chi2_contingency(contingency_table)

chi2, p_value
```

#### Interpretation

```python
# Statistical Result:
 # The p-value obtained from the chi-square test is <0.3408848783236828>.

# Statistical Decision:
 # If p-value < 0.05, the null hypothesis is rejected.

# Business Interpretation:
 # Expense types vary by business unit, indicating differences in operational cost structures.

# Confidence Statement: 
 # The result provides confidence that the relationship is statistically meaningful.
```

### Hypothesis 4: Cost per Employee Across Departments (ANOVA)

```python
# Business Question
 # Do different business units have significantly different average cost per employee?

#Metric Used
 # Cost per Employee (cost_to_company)

# Grouping Variable
 # business_unit

# Hypotheses
 # H₀: The average cost per employee is the same across all business units.
 # H₁: At least one business unit has a different average cost per employee.

# Statistical Test Used
 # ANOVA (used because more than two business units are compared)
```

```python
employee_groups = [
    headcount_df[headcount_df["business_unit"] == unit]["cost_to_company"]
    for unit in headcount_df["business_unit"].dropna().unique()
]

f_stat, p_value = f_oneway(*employee_groups)

f_stat, p_value
```

#### Interpretation

```python
# Statistical Result:
 # The p-value obtained from the ANOVA test is <0.6951671842922937>.

# Statistical Decision:
 # If p-value < 0.05, the null hypothesis is rejected.

# Business Interpretation:
 # Employee costs differ significantly across business units, indicating variations in compensation structures, role requirements, or workforce composition.

# Confidence Statement:  
 # The statistical test provides sufficient confidence in the observed differences.
```
