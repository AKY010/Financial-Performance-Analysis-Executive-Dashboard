```python
import pandas as pd 
import pymysql
import os
import warnings
import mysql.connector
warnings.filterwarnings('ignore')
```

```python
# Password is stored securely using environment variable
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password=os.getenv("DB_PASSWORD"),
    database="financial_analysis_db"
)
```

```python
budget_df = pd.read_sql("SELECT * FROM budget", conn)
print("Budgets:", budget_df)
```

```python
customers_df = pd.read_sql("SELECT * FROM customers", conn)
print("Customers:", customers_df)
```

```python
financial_transactions_df = pd.read_sql("SELECT * FROM financial_transactions", conn)
print("Financial_Transactions:", financial_transactions_df)
```

```python
headcount_df = pd.read_sql("SELECT * FROM headcount", conn)
print("Headcounts:", headcount_df)
```

```python
vendors_df = pd.read_sql("SELECT * FROM vendors", conn)
print("Vendors:", vendors_df)
```

```python
conn.close()
```
